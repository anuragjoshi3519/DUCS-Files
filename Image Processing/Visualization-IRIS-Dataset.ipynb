```python
from sklearn.datasets import load_iris
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# loading IRIS Dataset
iris = load_iris()
X_data = iris.data     # features
Y_data = iris.target   # classes
iris_data = np.hstack((X_data,Y_data.reshape(-1,1)))

# Creating Dataframe out of the numpy array
dfIris = pd.DataFrame(iris_data)
dfIris.columns = ['Sepal Length','Sepal Width','Petal Length','Petal Width','Class']
dfIris['Class'] = [iris.target_names[i] for i in list(np.array(dfIris['Class'],dtype='int16'))]

# visualising data'Class'])]
dfIris.head()
```

## Making visualisations using histograms

```python
i=1
plt.figure(figsize=(14,14)).suptitle('IRIS Frequency Histogram',fontsize=16)
for col_name in dfIris.columns[:-1]:
    plt.subplot(2,2,i)
    plt.hist(dfIris.loc[dfIris['Class']=='setosa',col_name],bins=12,color='red',label='Setosa',edgecolor='black')
    plt.hist(dfIris.loc[dfIris['Class']=='versicolor',col_name],bins=12,color='green',label='Versicolor',edgecolor='black')
    plt.hist(dfIris.loc[dfIris['Class']=='virginica',col_name],bins=12,color='blue',label='Virginica',edgecolor='black')
    plt.ylabel('Frequency',fontsize=14)
    plt.xlabel(col_name+' (cm)',fontsize=14)
    plt.legend()
    i+=1
plt.show()
```

### Making visualisations using Scatter Plots

```python
plt.figure(figsize=(16,6))
plt.subplot(1,2,1)
plt.scatter(iris_data[:,0],iris_data[:,2],c=iris_data[:,4])
plt.title('Sepal-Length Vs Petal-Length')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')

plt.subplot(1,2,2)
plt.scatter(iris_data[:,1],iris_data[:,3],c=iris_data[:,4])
plt.title('Sepal-Width Vs Petal-Width')
plt.xlabel('Sepal Width')
plt.ylabel('Petal Width')
plt.show()
```
